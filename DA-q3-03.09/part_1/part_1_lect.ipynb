```python
import pandas as pd
import matplotlib.pyplot as plt

# Создадим простую таблицу с данными о продажах
data = {"Месяц": ["Янв", "Фев", "Мар", "Апр"], "Продажи": [10, 14, 12, 18]}
df = pd.DataFrame(data)
print("Наша таблица:\n", df)

# Посчитаем простую статистику по столбцу "Продажи"
print("\nСтатистика по продажам:\n", df["Продажи"].describe())

# Построим график продаж по месяцам
df.plot(x="Месяц", y="Продажи", kind="line", marker="o", title="Продажи по месяцам")
plt.tight_layout()
plt.show()
```

```python
# Создание Series из списка
s = pd.Series([10, 20, 30, 40], name="Продажи")
print("Series:\n", s)

# Индекс и значения
print("\nИндекс:", s.index)     # метки строк (по умолчанию от 0 до n-1)
print("Значения:", s.values)   # сами данные (массив NumPy)

# Обращение по индексу и по номеру
print("\nПо индексу 2:", s[2])         # значение на позиции 2
print("По срезу [1:3]:\n", s[1:3])    # подмножество значений

# Метаданные
print("\nИмя Series:", s.name)
print("Тип данных:", s.dtype)
```

```python
# Создание DataFrame из словаря
data = {
    "Категория": ["A", "A", "B", "B", "C"],
    "Продажи": [10, 15, 20, 25, 30],
    "Город": ["Москва", "СПб", "Москва", "СПб", "Казань"]
}
df = pd.DataFrame(data)
print("Таблица DataFrame:\n", df)

# Выборка столбца и строки
print("\nСтолбец 'Продажи':\n", df["Продажи"])
print("\nПервая строка (iloc[0]):\n", df.iloc[0])

# Группировка и агрегирование
grouped = df.groupby("Категория")["Продажи"].mean()
print("\nСредние продажи по категориям:\n", grouped)

# Описательные статистики по числовым данным
print("\nСтатистики:\n", df.describe())
```

```python
# Создадим таблицу
data = {
    "Категория": ["A", "A", "B", "B", "C", "C"],
    "Город": ["Москва", "СПб", "Москва", "СПб", "Москва", "СПб"],
    "Продажи": [10, 15, 20, 25, 30, 35]
}
df = pd.DataFrame(data)
print("Исходная таблица:\n", df)

# Сводная таблица: средние продажи по категориям и городам
pivot = pd.pivot_table(df, values="Продажи", index="Категория", columns="Город", aggfunc="mean")
print("\nСводная таблица:\n", pivot)

# Добавим ещё статистику: суммы продаж по городам
pivot_sum = pd.pivot_table(df, values="Продажи", index="Город", aggfunc="sum")
print("\nСуммарные продажи по городам:\n", pivot_sum)

# Быстрый просмотр корреляций между столбцами
print("\nКорреляции:\n", df.corr(numeric_only=True))
```

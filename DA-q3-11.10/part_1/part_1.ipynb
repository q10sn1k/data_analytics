```python
import matplotlib.pyplot as plt
```

```python
# данные
x = [i for i in range(6)]
y1 = [i for i in x]
y2 = [i**2 for i in x]

fig, ax = plt.subplots()

ax.plot(x, y1, label="y = x")
ax.plot(x, y2, label="y = x**2")

ax.set_title("main is Figure/Axes")
ax.set_xlabel("ось X")
ax.set_ylabel("ось y")
ax.grid(True)
ax.annotate("выделим точку", xy=(2, 4), xytext=(3, 5),
            arrowprops=dict(arrowstyle="->"))

plt.show()
```

```python
products = ("Ноутбук", "Смартфон", "Наушники", "монитор",
           "клавиатура", "Мышь", "Планшет", "Планшет") # товары (Ось Y)

channels = 4 # число сегментов (каналы сбыта)

data = [
    [15.0, 22.0, 10.0, 18.0, 9.0, 12.0, 8.0, 14.0], # онлайн
    [12.0, 8.0, 11.0, 15.0, 10.0, 14.0, 9.0, 7.0],  # Розница
    [7.0, 5.0, 9.5, 6.0, 12.5, 6.5, 10.0, 8.0],     # Партнеры
    [5.5, 6.5, 7.0, 9.0, 8.0, 7.0, 6.0, 5.5]        # Опт  
]

data = np.array(data)

print(data)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

bar_height = 0.5
opacity = 0.85
ind = np.arange(len(products))

legend_labels = ["Онлайн","Розница", "Партнеры", "Опт"]

for i in range(channels):
    if i == 0:
        rects = plt.barh(ind, data[i], height=bar_height, alpha=opacity, label=legend_labels[i])
    else:
        rects = plt.barh(
            ind,
            data[i],
            height=bar_height,
            alpha=opacity,
            left=data[:i].sum(axis=0),
            label=legend_labels[i]
        )
    for rect in rects:
        width = rect.get_width()
        cx = rect.get_x() + width / 2.0      
        cy = rect.get_y() + rect.get_height() / 2.0
        
        ax.annotate(
            f"{width:.1f}",
            xy=(cx, cy),
            xytext=(0, -10),
            textcoords="offset points",
            ha="center", va="bottom",
            fontsize=8, color="black"
        )



ax.set_title("Продажи по товарам и каналам")
ax.set_xlabel("Объем продаж")
ax.set_ylabel("Товары")
ax.grid(axis="x", linestyle=":", alpha=0.5)

plt.yticks(ind, products)

plt.legend(loc="upper right")

plt.tight_layout()

plt.show()
```

```python
# подробно

import numpy as np
import matplotlib.pyplot as plt

# ------------------------ ДАННЫЕ ------------------------
# Кортеж названий товаров. Он пойдёт на ось Y как подписи к строкам.
# Обратите внимание: здесь есть два одинаковых элемента "Планшет" — это допустимо,
# просто на диаграмме будут две строки с одинаковой подписью.
products = ("Ноутбук", "Смартфон", "Наушники", "монитор",
            "клавиатура", "Мышь", "Планшет", "Планшет")  # товары (ось Y)

channels = 4  # число сегментов (каналов сбыта) внутри каждой горизонтальной полосы

# Матрица продаж: каждая ВНУТРЕННЯЯ список — это ОТДЕЛЬНЫЙ КАНАЛ (строка),
# каждый элемент внутри — объём продаж для конкретного товара.
# Т.е. итоговая форма должна быть (channels, n_products).
data = [
    [15.0, 22.0, 10.0, 18.0,  9.0, 12.0,  8.0, 14.0],  # Онлайн
    [12.0,  8.0, 11.0, 15.0, 10.0, 14.0,  9.0,  7.0],  # Розница
    [ 7.0,  5.0,  9.5,  6.0, 12.5,  6.5, 10.0,  8.0],  # Партнеры
    [ 5.5,  6.5,  7.0,  9.0,  8.0,  7.0,  6.0,  5.5],  # Опт
]

# Превращаем список списков в NumPy-массив, чтобы было удобно делать векторные операции,
# например, суммировать данные по каналам (ось=0).
data = np.array(data, dtype=float)

# (Необязательно) sanity-check: убеждаемся, что размеры согласованы.
assert data.shape[0] == channels, "Число строк в data должно равняться числу каналов"
assert data.shape[1] == len(products), "Число столбцов в data должно равняться числу товаров"

print("Форма data:", data.shape)  # (channels, n_products)
print(data)

# ------------------------ ФИГУРА И ОСИ ------------------------
# plt.subplots возвращает пару (fig, ax):
# - fig: объект Figure — «полотно» для отрисовки
# - ax:  объект Axes  — система координат, куда мы будем рисовать графики
# figsize=(10, 5) задаёт размер рисунка в дюймах: ширина=10, высота=5
fig, ax = plt.subplots(figsize=(10, 5))

# Параметры внешнего вида столбиков:
bar_height = 0.5  # толщина горизонтальной полосы (в единицах координат оси Y)
opacity = 0.85    # прозрачность (alpha): 1.0 — непрозрачно, 0 — прозрачно

# Индексы по оси Y для размещения строк (0..N-1).
# np.arange(n) создаёт массив [0,1,2,...,n-1]
ind = np.arange(len(products))

# Подписи для легенды — по одному на канал (порядок соответствует строкам массива data).
legend_labels = ["Онлайн", "Розница", "Партнеры", "Опт"]

# ------------------------ ОСНОВНОЙ ЦИКЛ ПО КАНАЛАМ ------------------------
# Идея стековой диаграммы:
# - Для i=0 рисуем бары от нуля (левый край по X = 0).
# - Для i>0 указываем аргумент left=..., чтобы сдвинуть новый сегмент вправо
#   на сумму всех предыдущих слоёв (каналов) — это и есть «stacked».
for i in range(channels):
    if i == 0:
        # plt.barh — рисует горизонтальные столбики.
        # Аргументы:
        #   ind           — координаты центров полос по оси Y (каждый товар — своя строка)
        #   data[i]       — ширины полос по оси X (значения для КАНАЛА i)
        #   height        — толщина полос
        #   alpha         — прозрачность
        #   label         — подпись в легенду (для этого слоя/канала)
        rects = plt.barh(
            ind,
            data[i],
            height=bar_height,
            alpha=opacity,
            label=legend_labels[i]
        )
    else:
        # left=data[:i].sum(axis=0):
        #   data[:i] — берём все предыдущие каналы (0..i-1),
        #   sum(axis=0) — суммируем по каналам для каждого товара,
        #   получаем «накопленную базу», от которой начнётся текущий сегмент.
        rects = plt.barh(
            ind,
            data[i],
            height=bar_height,
            alpha=opacity,
            left=data[:i].sum(axis=0),
            label=legend_labels[i]
        )

    # ------------------------ АННОТАЦИИ НА СЕГМЕНТАХ ------------------------
    # rects — контейнер (BarContainer) с прямоугольниками (Rectangle) для текущего канала.
    # Пройдём по каждому прямоугольнику, чтобы подписать его значение внутри.
    for rect in rects:
        # rect.get_width()  — ширина сегмента по оси X (текущее числовое значение)
        width = rect.get_width()
        # rect.get_x()      — левый X сегмента; половина ширины даёт центр по X
        cx = rect.get_x() + width / 2.0
        # rect.get_y()      — нижняя граница сегмента; половина высоты даёт центр по Y
        cy = rect.get_y() + rect.get_height() / 2.0

        # ax.annotate — рисует текстовую аннотацию.
        # Аргументы:
        #   text=f"{width:.1f}"      — что писать (число с одним знаком после запятой)
        #   xy=(cx, cy)              — точка привязки (центр сегмента)
        #   xytext=(0, -10)          — смещение текста относительно точки привязки в пунктах
        #   textcoords="offset points" — говорим, что xytext интерпретируется как смещение в «points»
        #   ha="center", va="bottom" — выравнивание текста по горизонтали/вертикали
        #   fontsize=8, color="black" — оформление текста
        ax.annotate(
            f"{width:.1f}",
            xy=(cx, cy),
            xytext=(0, -10),
            textcoords="offset points",
            ha="center", va="bottom",
            fontsize=8, color="black"
        )

# ------------------------ ОФОРМЛЕНИЕ ГРАФИКА ------------------------
# Заголовок и подписи осей — контекст для читаемости графика.
ax.set_title("Продажи по товарам и каналам")
ax.set_xlabel("Объем продаж")
ax.set_ylabel("Товары")

# Сетка по оси X помогает визуально считывать значения:
# axis="x" — рисуем вертикальные линии сетки, соответствующие делениям по X,
# linestyle=":" — пунктир, чтобы сетка была ненавязчивой,
# alpha=0.5 — полупрозрачность.
ax.grid(axis="x", linestyle=":", alpha=0.5)

# Подменяем числовые индексы ind на человекочитаемые подписи товаров.
# plt.yticks(позиции, подписи)
plt.yticks(ind, products)

# Легенда показывает соответствие цвета/слоя — каналу продаж.
# loc="upper right" — расположение в правом верхнем углу области построения.
plt.legend(loc="upper right")

# Автоподбор отступов, чтобы подписи и легенда не обрезались рамкой фигуры.
plt.tight_layout()

# Отобразить график.
plt.show()
```
